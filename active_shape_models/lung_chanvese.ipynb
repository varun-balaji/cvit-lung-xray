```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import skimage
from skimage.segmentation import chan_vese
%matplotlib inline
```

```python
lung = cv2.imread('/home/varun/cvit/lung_xray/resources/Chest xray dataset/annotated images/00001024_000.png', 0)
plt.imshow(lung, cmap='gray')
```

```python
mu=0.25 
lambda1=1
lambda2=1
tol=1e-3
max_iter=200
dt=0.5
init_level_set="checkerboard"
extended_output=True

cv = chan_vese(lung, mu, lambda1, lambda2, tol, max_iter, dt, init_level_set, extended_output)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(8, 8))
ax = axes.flatten()

ax[0].imshow(lung, cmap="gray")
ax[0].set_axis_off()
ax[0].set_title("Original Image", fontsize=12)

ax[1].imshow(cv[0], cmap="gray")
ax[1].set_axis_off()
title = "Chan-Vese segmentation - {} iterations".format(len(cv[2]))
ax[1].set_title(title, fontsize=12)

ax[2].imshow(cv[1], cmap="gray")
ax[2].set_axis_off()
ax[2].set_title("Final Level Set", fontsize=12)

ax[3].plot(cv[2])
ax[3].set_title("Evolution of energy over iterations", fontsize=12)

fig.tight_layout()
plt.show()
```
