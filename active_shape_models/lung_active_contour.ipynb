```python
lung = cv2.imread('/home/varun/cvit/lung_xray/resources/Chest xray dataset/annotated images/00001024_000.png', 0)
plt.imshow(lung)
```

```python
# Initialize snake
s = np.linspace(0, 2*np.pi, 400)
x = 700 + 200*np.cos(s)
y = 600 + 400*np.sin(s)
init = np.array([x, y]).T
```

```python
# No parameters
snake = active_contour(skimage.filters.gaussian(lung, 3), init)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
plt.gray()
ax.imshow(lung)
ax.plot(init[:, 0], init[:, 1], 'r', lw=3)
ax.plot(snake[:, 0], snake[:, 1], 'b', lw=3)
```

```python
# Attract snake to edges
snake = active_contour(skimage.filters.gaussian(lung, 3), init, w_edge=10)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
plt.gray()
ax.imshow(lung)
ax.plot(init[:, 0], init[:, 1], 'r', lw=3)
ax.plot(snake[:, 0], snake[:, 1], 'b', lw=3)
```

```python
# Attract snake to dark regions
snake = active_contour(skimage.filters.gaussian(lung, 3), init, w_edge=10, w_line=-5)
```

```python
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
plt.gray()
ax.imshow(lung)
ax.plot(init[:, 0], init[:, 1], 'r', lw=3)
ax.plot(snake[:, 0], snake[:, 1], 'b', lw=3)
```
